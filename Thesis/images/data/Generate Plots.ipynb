```python
import pandas as pd
import numpy as np

from plotnine import *
from plotnine.data import *

%matplotlib inline
```

```python
def gemerate_plot(filename):
    df = pd.DataFrame(columns={"Batches","Perplexity", "Curriculum"})
    for row in pd.read_csv(filename).iterrows():
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][1],"Curriculum":"Baseline"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][2],"Curriculum":"Random"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][3],"Curriculum":"DEP"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][4],"Curriculum":"POS"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][5],"Curriculum":"Unigram"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][6],"Curriculum":"Bigram"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][7],"Curriculum":"Trigram"},ignore_index=True)
        df =df.append({"Batches":row[1][0],"Perplexity":row[1][8],"Curriculum":"Length"},ignore_index=True)
    return (ggplot(df, aes(x='Batches', y='Perplexity', color='Curriculum'))
         + geom_smooth(method='auto', span=0.0, se=False, level=0.00)
         # Then you can change the colour using a scale
        )    
```

```python
gemerate_plot('wikitext2sentence.csv')
```

```python
gemerate_plot('wikitext2line.csv')
```

```python
gemerate_plot('wikitext103line.csv')
```
